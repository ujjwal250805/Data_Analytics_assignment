#### Que:3 Write a query to calculate:
• Total number of employees

• Average salary

• Minimum and maximum salary

Explain the difference between aggregate and scalar functions.

 *Answer:* 
 
 Aggregate Vs Scalar function :
 
An aggregate function operates on multiple rows at the same time and returns a single result. A scalar function operates on one row at a time and returns one result for each row.

Generally, aggregate functions are used on numeric values (for example, SUM(), AVG(), COUNT()), while scalar functions are commonly used on character data such as VARCHAR to perform operations like UPPER().
