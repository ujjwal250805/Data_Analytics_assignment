1. Explain: Data Cleaning & Preprocessing

• Load the dataset and check for missing values.

• Fill missing Age values with median or mean.

• Handle missing Embarked values by filling with the most common port.

• Drop unnecessary columns like Name, Ticket, Cabin for visualization purposes


```python
import pandas as pd

df = pd.read_csv("C:\\Users\\DELL\\Downloads\\titanic.csv")

print(df.head())
print(df.isnull().sum())
```

Step 2:- Fill missing Age values with median

```python
median_age = df["Age"].median()
df["Age"] = df["Age"].fillna(median_age)
```

Step 3:- Handle missing Embarked values

```python
most_common_port = df["Embarked"].mode()[0]
df["Embarked"] = df["Embarked"].fillna(most_common_port)
```

Step 4:- Drop unnecessary columns


```python
df_cleaned = df.drop(columns=["Name", "Ticket", "Cabin"])
print(df_cleaned.head())
```

2. Univariate Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sn
```

Distribution of Age using histogram and KDE

```python
plt.figure()
sns.histplot(df_cleaned["Age"], kde=True)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

Count of passengers by Sex


```python
plt.figure()
sns.countplot(x="Sex", data=df_cleaned)
plt.title("Passenger Count by Sex")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.show()
```

Distribution of Fare using boxplot

```python
plt.figure()
sns.boxplot(y="Fare", data=df_cleaned)
plt.title("Fare Distribution")
plt.ylabel("Fare")
plt.show()
```
